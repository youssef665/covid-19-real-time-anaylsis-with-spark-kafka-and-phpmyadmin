```python
from pyspark.ml.feature import VectorAssembler
```

```python
dfspark = spark.read.csv("./modified.json",header=True,inferSchema=True)
```

```python
dfspark.show()
```

```python
featureassembler = VectorAssembler(inputCols=["new_cases","total_cases"],outputCol="Independent_feature")
```

```python
output =featureassembler.transform(dfspark)
```

```python
output.show()
```

```python
finalized_data = output.select("Independent_feature","total_deaths")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
train_data, test_data = finalized_data.randomSplit([0.75,0.25])
```

```python
regressor = LinearRegression(featuresCol='Independent_feature',labelCol='total_deaths')
```

```python
model  =  regressor.fit(train_data)
```

```python
pred_results = model.evaluate(test_data)
```

```python
expec = pred_results.predictions
expec.show()
```

```python
df12 = expec.drop('Independent_feature')
```

```python
df12.show()
```

```python
df12.toJSON("D:/big data project file/Big-data/Big-Data-files-main/two.json")
```
